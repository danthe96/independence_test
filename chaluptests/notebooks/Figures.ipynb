```python
import sys
sys.path.append('../../')

import numpy as np
from scipy import stats
import matplotlib as mpl
mpl.rc('figure', dpi=600)
mpl.rc('savefig', dpi=600)
mpl.rc('figure', facecolor='white')
%matplotlib inline

from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties

from collections import defaultdict
from independence_test.data import toy
```

```python
""" Plot HYBRID in 1D. """

plt.figure(figsize=(7.5, 7.5), facecolor='white')

np.random.seed(11)
x, y, z = toy.make_discrete_data(n_samples=1000, complexity=8, dim=2, type='indep')
z = z[:, 0]
z_ids = np.argsort(z)

z = z[z_ids]
x = x[z_ids, :8]
y = y[z_ids, :8]

ax = plt.subplot(2, 2, 1)
plt.title('hybrid (trials = 8, k = 2) indep')
plt.xlabel('')
plt.ylabel('z (multinomial distr. parameter)')
plt.imshow(x, interpolation='nearest', cmap='Greys', aspect='auto', origin='low')
plt.xticks(range(8), range(1, 9))
plt.yticks([0, 500, 1000], [0, .5, 1])

ax=plt.subplot(2, 2, 2)
plt.title('hybrid (trials = 8, k = 2) indep')
plt.xlabel('')
plt.ylabel('')
plt.imshow(y, interpolation='nearest', cmap='Greys', aspect='auto', origin='low')
plt.xticks(range(8), range(1, 9))
plt.yticks([0, 500, 1000], [])

x, y, z = toy.make_discrete_data(n_samples=1000, complexity=8, dim=2, type='dep')
z = z[:, 0]
z_ids = np.argsort(z)

z = z[z_ids]
x = x[z_ids, :8]
y = y[z_ids, :8]

ax = plt.subplot(2, 2, 3)
plt.title('hybrid (trials = 8, k = 2) dep')
plt.xlabel('x (times ''1'' was drawn)')
plt.ylabel('z (multinomial distr. parameter)')
plt.imshow(x, interpolation='nearest', cmap='Greys', aspect='auto', origin='low')
plt.xticks(range(8), range(1, 9))
plt.yticks([0, 500, 1000], [0, .5, 1])

ax=plt.subplot(2, 2, 4)
plt.title('hybrid (trials = 8, k = 2) dep')
plt.ylabel('')
plt.xlabel('x (times ''1'' was drawn)')
plt.imshow(y, interpolation='nearest', cmap='Greys', aspect='auto', origin='low')
plt.xticks(range(8), range(1, 9))
plt.yticks([0, 500, 1000], [])
```

```python
""" Plot Lingauss in 1D. """

plt.figure(figsize=(7.5, 7.5), facecolor='white')

np.random.seed(5)
x, y, z = toy.make_chain_data(n_samples=1000, complexity=1, dim=1, type='indep')
ax = plt.subplot(2, 2, 1)
plt.title('lingauss indep (x <- z -> y)')
plt.xlabel('')
plt.ylabel('y')
plt.plot(x, y, 'k.', alpha=.3)
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))


ax=plt.subplot(2, 2, 2)
plt.title('lingauss indep (x <- z -> y)')
plt.xlabel('')
plt.ylabel('')
plt.plot(z, y, 'k.', alpha=.3)
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))

x, y, z = toy.make_pnl_data(n_samples=1000, complexity=1, dim=1, type='dep')
ax=plt.subplot(2, 2, 3)
plt.title('lingauss dep (z -> x -> y)')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, 'k.', alpha=.3)
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))

ax=plt.subplot(2, 2, 4)
plt.title('lingauss dep (z -> x -> y)')
plt.xlabel('z')
plt.ylabel('')
plt.plot(z, y, 'k.', alpha=.3)
ax.yaxis.set_major_locator(plt.MaxNLocator(4))
ax.xaxis.set_major_locator(plt.MaxNLocator(4))
```
